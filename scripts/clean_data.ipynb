# Compile and clean EU data

```python
from functools import reduce

import numpy as np
import pandas as pd
```

### Clean original df

```python
df = pd.read_csv('../data/raw_data/eu_da_prices.csv', delimiter=';')
df
```

```python
df.head()
```

```python
df.columns = [
    'date', 'start_time', 'end_time', 'Germany', 'De-Lu', 'Belgium', 'Denmark 1',
    'Denmark 2', 'France', 'Netherlands', 'Norway 2', 'Austria', 'Poland',
    'Sweden 4', 'Switzerland', 'Czech', 'De-At-Lu', 'Italy', 'Slovenia', 'Hungary'
]
df.head()
```

```python
df.insert(1, 'datetime', pd.to_datetime(df.date + ' ' +  df.start_time))
df.date = pd.to_datetime(df.date)
df.drop(['start_time', 'end_time', 'De-At-Lu'], axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
df.replace('-', np.NaN, inplace=True)
df.replace(',', '', regex=True, inplace=True)
df['De-Lu'] = df['De-Lu'].astype(float)
df.France = df.France.astype(float)
df.Poland = df.Poland.astype(float)
df.Hungary = df.Hungary.astype(float)
df.info()
```

### Merge in GB and Bulgaria 2022 comparison in EUR

```python
bulgaria = pd.read_csv('../data/raw_data/bulagria_da.csv')
bulgaria.rename({'MTU (CET/CEST)': 'date', 'Day-ahead Price [Currency/MWh]': 'Bulgaria_lev'}, axis=1, inplace=True)
bulgaria.insert(1, 'hour', bulgaria.date.str[-5:-3].replace('00', '24').astype(int))
bulgaria.date = pd.to_datetime(bulgaria.date.str[:10], dayfirst=True)
bulgaria = bulgaria[['date', 'hour', 'Bulgaria_lev']]
bulgaria
```

```python
conversion_df
```

```python
epex_price_df = pd.read_csv('../data/raw_data/EPEX_and_N2EX_DA_WD_PRICES.csv')[['SETTLEMENT_DATE', 'SETTLEMENT_PERIOD', 'EPEX_DA']]
epex_price_df.columns = ['date', 'sp', 'gb_price_gbp']
epex_price_df['hour'] = np.ceil(epex_price_df.sp / 2).astype(int)
epex_price_df.drop('sp', axis=1, inplace=True)
epex_price_df.drop_duplicates(subset=['date', 'hour'], inplace=True)
epex_price_df.date = pd.to_datetime(epex_price_df.date)
epex_price_df = pd.merge(epex_price_df, bulgaria, on=['date', 'hour'], how='inner')

conversion_df = pd.read_csv('../data/raw_data/eur_conversion.csv')
conversion_df.date = pd.to_datetime(conversion_df.date)
conversion_df = conversion_df[['date', 'Pound sterling', '(Bulgaria) Lev']]
epex_price_df = pd.merge(epex_price_df, conversion_df, on='date', how='left')
epex_price_df['GB'] = epex_price_df.gb_price_gbp / epex_price_df['Pound sterling']
epex_price_df['Bulgaria'] = epex_price_df.Bulgaria_lev / epex_price_df['(Bulgaria) Lev']
epex_price_df = epex_price_df.round(2)
epex_price_df.drop(['gb_price_gbp', 'Pound sterling', '(Bulgaria) Lev', 'Bulgaria_lev'], axis=1, inplace=True)
epex_price_df
```

### Merge in data from Entsoe

```python
greece = pd.read_csv('../data/raw_data/greek_da.csv').drop(['Currency', 'BZN|GR'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
croatia = pd.read_csv('../data/raw_data/croatia_da.csv').drop(['Currency', 'BZN|HR'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
finland = pd.read_csv('../data/raw_data/finland_da.csv').drop(['Currency', 'BZN|FI'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
ireland = pd.read_csv('../data/raw_data/ireland_da.csv').drop(['Currency', 'BZN|IE(SEM)'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
romania = pd.read_csv('../data/raw_data/romania_da.csv').drop(['Currency', 'BZN|RO'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
spain = pd.read_csv('../data/raw_data/spain_da.csv').drop(['Currency', 'BZN|ES'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
serbia = pd.read_csv('../data/raw_data/serbia_da.csv').drop(['Currency', 'BZN|RS'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
sweden_1 = pd.read_csv('../data/raw_data/sweden_1_da.csv').drop(['Currency', 'BZN|SE1'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
sweden_2 = pd.read_csv('../data/raw_data/sweden_2_da.csv').drop(['Currency', 'BZN|SE2'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
sweden_3 = pd.read_csv('../data/raw_data/sweden_3_da.csv').drop(['Currency', 'BZN|SE3'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
norway_1 = pd.read_csv('../data/raw_data/norway_1_da.csv').drop(['Currency', 'BZN|NO1'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
norway_3 = pd.read_csv('../data/raw_data/norway_3_da.csv').drop(['Currency', 'BZN|NO3'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
norway_4 = pd.read_csv('../data/raw_data/norway_4_da.csv').drop(['Currency', 'BZN|NO4'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
norway_5 = pd.read_csv('../data/raw_data/norway_5_da.csv').drop(['Currency', 'BZN|NO5'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
estonia = pd.read_csv('../data/raw_data/estonia_da.csv').drop(['Currency', 'BZN|EE'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
latvia = pd.read_csv('../data/raw_data/latvia_da.csv').drop(['Currency', 'BZN|LV'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
lithuania = pd.read_csv('../data/raw_data/lithuania_da.csv').drop(['Currency', 'BZN|LT'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')
portugal = pd.read_csv('../data/raw_data/portugal_da.csv').drop(['Currency', 'BZN|PT'], axis=1).drop_duplicates(subset='MTU (CET/CEST)')

greece.rename({'Day-ahead Price [EUR/MWh]': 'Greece'}, axis=1, inplace=True)
croatia.rename({'Day-ahead Price [EUR/MWh]': 'Croatia'}, axis=1, inplace=True)
finland.rename({'Day-ahead Price [EUR/MWh]': 'Finland'}, axis=1, inplace=True)
ireland.rename({'Day-ahead Price [EUR/MWh]': 'Ireland'}, axis=1, inplace=True)
romania.rename({'Day-ahead Price [EUR/MWh]': 'Romania'}, axis=1, inplace=True)
spain.rename({'Day-ahead Price [EUR/MWh]': 'Spain'}, axis=1, inplace=True)
serbia.rename({'Day-ahead Price [EUR/MWh]': 'Serbia'}, axis=1, inplace=True)
sweden_1.rename({'Day-ahead Price [EUR/MWh]': 'Sweden 1'}, axis=1, inplace=True)
sweden_2.rename({'Day-ahead Price [EUR/MWh]': 'Sweden 2'}, axis=1, inplace=True)
sweden_3.rename({'Day-ahead Price [EUR/MWh]': 'Sweden 3'}, axis=1, inplace=True)
norway_1.rename({'Day-ahead Price [EUR/MWh]': 'Norway 1'}, axis=1, inplace=True)
norway_3.rename({'Day-ahead Price [EUR/MWh]': 'Norway 3'}, axis=1, inplace=True)
norway_4.rename({'Day-ahead Price [EUR/MWh]': 'Norway 4'}, axis=1, inplace=True)
norway_5.rename({'Day-ahead Price [EUR/MWh]': 'Norway 5'}, axis=1, inplace=True)
estonia.rename({'Day-ahead Price [EUR/MWh]': 'Estonia'}, axis=1, inplace=True)
latvia.rename({'Day-ahead Price [EUR/MWh]': 'Latvia'}, axis=1, inplace=True)
lithuania.rename({'Day-ahead Price [EUR/MWh]': 'Lithuania'}, axis=1, inplace=True)
portugal.rename({'Day-ahead Price [EUR/MWh]': 'Portugal'}, axis=1, inplace=True)

df_list = [
    greece, croatia, finland, ireland, romania, spain, serbia, sweden_1, sweden_2, sweden_3,
    norway_1, norway_3, norway_4, norway_5, estonia, latvia, lithuania, portugal
    ]

added_countries = reduce(lambda left, right: pd.merge(left, right, on='MTU (CET/CEST)', how='inner'), df_list)
added_countries.rename({'MTU (CET/CEST)': 'date'}, axis=1, inplace=True)
added_countries.insert(1, 'hour', added_countries.date.str[-5:-3].replace('00', '24').astype(int))
added_countries.date = pd.to_datetime(added_countries.date.str[:10], dayfirst=True)
added_countries
```

```python
df = pd.merge(df, added_countries, on=['date', 'hour'], how='left')
df.sort_index(axis=1, inplace=True)
df.tail()
```

```python
df.to_csv('../data/cleaned_data/eu_da_prices_clean.csv')
df
```
